# Most expensive and cheapest wines per country

```python
#> include average price on each bar

import pandas as pd
import matplotlib.pyplot as plt

# data
data = pd.read_csv('/Users/dilarabursa/Desktop/TechLabsDUS_WineNot/data/raw/winemag-data-130k-v2.csv')

# filtering data for "country" and "price" 
filtered_data = data[['country', 'price']]

# average price by country
average_price_by_country = filtered_data.groupby('country')['price'].mean()

# sort by average price
sorted_countries = average_price_by_country.sort_values(ascending=False)

# most 10 expensive countries
top_10_expensive = sorted_countries.head(10)

# most 10 cheapest countries
top_10_cheapest = sorted_countries.tail(10)

# Plotten
#plt.figure(figsize=(15, 8))
#plt.bar(top_10_expensive.index, top_10_expensive.values, color='red',, label='expensive country')
#plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='cheapest country')
#plt.xlabel('Country')
#plt.ylabel('Average Price')
#plt.title('Most expensive and cheapest wines per country')
#plt.legend()
#plt.xticks(rotation=90)
#plt.show()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=10, color='black')
plt.figure(figsize=(18, 10))
autolabel(plt.bar(top_10_expensive.index, top_10_expensive.values, color='red', label='expensive country'))
autolabel(plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='cheapest country'))
plt.xlabel('Country')
plt.ylabel('Average Price')
plt.title('Most expensive and cheapest wines per country')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
