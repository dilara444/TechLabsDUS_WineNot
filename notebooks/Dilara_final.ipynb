```python
#Dilara new plot : Please check again I have problems with this one
#To plot how pricy the different types of wines are listed
#The "price" column occasionally contains N/A values, which is why they must be excluded for plotting. Otherwise the command cannot be finished. 
#Das muss beim Plotten auf jedenfall erwähnt werden!

df_wine1_filtered = df_wine1[df_wine1['price'].notnull()]
plt.figure(figsize=(10, 6))
sns.histplot(data=df_wine1_filtered, x='price', hue='variety', multiple='stack', palette='Set2', element='step')
plt.title("Histogram of prices by type of wine", fontsize=20)
plt.xlabel("Price (US-Dollar)", fontsize=15)
plt.ylabel("Frequency", fontsize=15)
plt.legend(title="Type of wine")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Erstellen Sie eine Liste der Weinsorten und ihrer Preise
wine_prices = df_wine1_filtered.groupby('variety')['price'].mean().reset_index()
wine_prices['variety'] = wine_prices['variety'].str.capitalize()  # Erste Buchstaben groß schreiben

# Erstellen Sie eine Wortwolke
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(zip(wine_prices['variety'], wine_prices['price'])))

# Zeigen Sie die Wortwolke an
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```

```python
#> include average price on each bar

import pandas as pd
import matplotlib.pyplot as plt

# Beispiel-Daten (zufällige Werte)
data = pd.read_csv('/Users/dilarabursa/Desktop/TechLabsDUS_WineNot/data/raw/winemag-data-130k-v2.csv')

# Daten filtern: Nur "country" und "price" behalten
filtered_data = data[['country', 'price']]

# Durchschnittspreis pro Land berechnen
average_price_by_country = filtered_data.groupby('country')['price'].mean()

# Sortieren nach Durchschnittspreis
sorted_countries = average_price_by_country.sort_values(ascending=False)

# Top 10 teuerste Länder
top_10_expensive = sorted_countries.head(10)

# Top 10 billigste Länder
top_10_cheapest = sorted_countries.tail(10)

# Plotten
#plt.figure(figsize=(15, 8))
#plt.bar(top_10_expensive.index, top_10_expensive.values, color='red', label='Teuerste Länder')
#plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='Billigste Länder')
#plt.xlabel('Country')
#plt.ylabel('Average Price')
#plt.title('Most expensive and cheapest wines per country')
#plt.legend()
#plt.xticks(rotation=90)
#plt.show()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=10, color='black')
plt.figure(figsize=(18, 10))
autolabel(plt.bar(top_10_expensive.index, top_10_expensive.values, color='red', label='expensive country'))
autolabel(plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='cheapest country'))
plt.xlabel('Country')
plt.ylabel('Average Price')
plt.title('Most expensive and cheapest wines per country')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
