# Scatterplot: Price vs. review points by country

This Scatterplots gives an overview about how the prices are for the quality of wines in each country listed in the dataset. This plot shows maily, that there is one outliner in the dataset, hwich is rated quite as average compared to other wines in this dataset, but still is sold very pricy in one of the countries. This scatterplots provides no clear distuingishing and conclusion about how much a wine costs per country nor the average quality of each country. It just gives a brief overview of how the prices in dependence of quality ratings are located globally in the world. It's important to note, that the values of used dataset are depriving from subjective rating and contain only partly datapoints of wine ratings all over the world. 

```python
#Dilara new plot : plotting by regions gives a very long legend and the plot is not übersichtlich
# This Scatterplots gives an overview about how the prices are for the quality of wines in each country listed in the dataset

plt.figure(figsize=(15, 10))
sns.scatterplot(x='price', y='points', hue='country', data=df_wine1, palette='viridis')
plt.title("Scatterplot: Price vs. review points by country", fontsize=20)
plt.xlabel("price (US-Dollar)", fontsize=15)
plt.ylabel("points", fontsize=15)
plt.legend(title="country", loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=6)
plt.show()
```

```python
#> include average price on each bar

import pandas as pd
import matplotlib.pyplot as plt

# Beispiel-Daten (zufällige Werte)
data = pd.read_csv('/Users/dilarabursa/Desktop/TechLabsDUS_WineNot/data/raw/winemag-data-130k-v2.csv')

# Daten filtern: Nur "country" und "price" behalten
filtered_data = data[['country', 'price']]

# Durchschnittspreis pro Land berechnen
average_price_by_country = filtered_data.groupby('country')['price'].mean()

# Sortieren nach Durchschnittspreis
sorted_countries = average_price_by_country.sort_values(ascending=False)

# Top 10 teuerste Länder
top_10_expensive = sorted_countries.head(10)

# Top 10 billigste Länder
top_10_cheapest = sorted_countries.tail(10)

# Plotten
#plt.figure(figsize=(15, 8))
#plt.bar(top_10_expensive.index, top_10_expensive.values, color='red', label='Teuerste Länder')
#plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='Billigste Länder')
#plt.xlabel('Country')
#plt.ylabel('Average Price')
#plt.title('Most expensive and cheapest wines per country')
#plt.legend()
#plt.xticks(rotation=90)
#plt.show()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=10, color='black')
plt.figure(figsize=(18, 10))
autolabel(plt.bar(top_10_expensive.index, top_10_expensive.values, color='red', label='expensive country'))
autolabel(plt.bar(top_10_cheapest.index, top_10_cheapest.values, color='green', label='cheapest country'))
plt.xlabel('Country')
plt.ylabel('Average Price')
plt.title('Most expensive and cheapest wines per country')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
